```python
x = df[df['supplier_okpo'] == '40581739']
x = x.drop(clear_cols, axis=1)
```

```python
y = x['winner']
```

```python
guarantees = [0 if isnan(x) else float(x) for x in x['guarantee']]
```

```python
x['guarantee']=guarantees
```

```python
x = x.drop(['organizer_region', 'organizer_city', 'organizer_okpo', 'date_publish_dt', 'winner'], axis=1)
```

```python
x.info()
```

```python
x = x.fillna(0)
```

```python
x = x.drop(['supplier_okpo', 'disqualified'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
preds = rfc.predict(X_test)
print('ROC AUC:', roc_auc_score(preds, y_test))
```

```python
import xgboost as xgb
xgbc = xgb.XGBClassifier()
xgbc.fit(X_train, y_train)
preds = xgbc.predict(X_test)
print('ROC AUC:', roc_auc_score(preds, y_test))
```
